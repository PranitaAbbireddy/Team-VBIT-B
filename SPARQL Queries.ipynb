```python
from SPARQLWrapper import SPARQLWrapper,JSON


# Define the SPARQL endpoint URL
endpoint_url = "https://query.wikidata.org/sparql"

# Create a SPARQLWrapper object
wrapper = SPARQLWrapper(endpoint_url)

# Define your SPARQL query
query = """
#men in entertainment( actor in movies, series, etc)
SELECT DISTINCT ?item ?itemLabel WHERE {
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE]". }
  {
    SELECT DISTINCT ?item WHERE {
      ?item p:P27 ?statement0.
      ?statement0 (ps:P27/(wdt:P279*)) wd:Q668.
      ?item p:P21 ?statement1.
      ?statement1 (ps:P21/(wdt:P279*)) wd:Q6581097.
      ?item p:P106 ?statement2.
      ?statement2 (ps:P106/(wdt:P279*)) wd:Q33999.
    }
  }
}"""

# Set the query string
wrapper.setQuery(query)

# Set the return format to JSON
wrapper.setReturnFormat(JSON)

# Execute the query and get the results
results = wrapper.query().convert()

# Print the results
print("Results:")
for item in results["results"]["bindings"]:
    print(item["itemLabel"]["value"])
```
